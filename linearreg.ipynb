## Visualize distributions

### Subtask:
Create histograms or other plots to visualize the distribution of each feature and the target variable.


**Reasoning**:
Iterate through each column of the dataframe and create a histogram for each to visualize the distribution.



```python
import matplotlib.pyplot as plt

for column in df_california.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df_california[column], bins=50)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

### Subtask:
Generate scatter plots or other visualizations to explore the relationships between features and the target variable.


**Reasoning**:
Iterate through the feature columns and create scatter plots against the target variable.



```python
features = df_california.columns.drop('MedHouseVal')

for feature in features:
    plt.figure(figsize=(8, 6))
    plt.scatter(df_california[feature], df_california['MedHouseVal'], alpha=0.5, s=10)
    plt.title(f'Relationship between {feature} and MedHouseVal')
    plt.xlabel(feature)
    plt.ylabel('Median House Value')
    plt.show()
```

## Check for multicollinearity

### Subtask:
Assess multicollinearity among features, which can affect the stability and interpretability of a linear regression model.


**Reasoning**:
Calculate the VIF for each feature and store it in a DataFrame, then sort and display the DataFrame.



```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

features = df_california.columns.drop('MedHouseVal')
X = df_california[features]

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data = vif_data.sort_values(by='VIF', ascending=False)

display(vif_data)
```

## Summary:

### Data Analysis Key Findings

*   Box plots for `AveRooms` and `AveOccup` show significant outliers, with values extending far beyond the whiskers.
*   Variance Inflation Factor (VIF) analysis revealed high multicollinearity between `Longitude` (VIF = 76.5) and `Latitude` (VIF = 80.9), as well as moderately high multicollinearity between `AveRooms` (VIF = 28.3) and `AveBedrms` (VIF = 25.4).

### Insights or Next Steps

*   Consider strategies like capping or removal to handle the identified outliers in `AveRooms` and `AveOccup` before building a linear regression model.
*   Address the high multicollinearity, potentially by removing one of the highly correlated features or using dimensionality reduction techniques, to ensure a stable and interpretable linear regression model.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(california.data,
                                                    california.target, random_state=11)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(X=X_train, y=y_train)

for i,name in enumerate(california.feature_names):
    print(f'{name:>10}: {linear_regression.coef_[i]}')
```

```python
print(linear_regression.intercept_)
```

```python
predicted = linear_regression.predict(X_test)
expected = y_test
```

```python
df = pd.DataFrame()
df['Expected'] = pd.Series(expected)
df['Predicted'] = pd.Series(predicted)

figure = plt.figure(figsize=(9, 9))
axes = sns.scatterplot(data=df, x='Expected', y='Predicted',
                       hue='Predicted', palette='cool', legend=False)
```

```python
start = min(expected.min(), predicted.min())
end = max(expected.max(), predicted.max())

print(f'X-limit {axes.set_xlim(start, end)}')
print(f'Y-limit {axes.set_ylim(start, end)}')
```

```python
line = plt.plot([start, end], [start, end], 'k--')
```

```python
from sklearn import metrics
print(f'The R2 score: {metrics.r2_score(expected, predicted)}\n')
print(f'The MSE: {metrics.mean_squared_error(expected, predicted)}')
```
